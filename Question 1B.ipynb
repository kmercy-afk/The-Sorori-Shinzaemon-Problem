```python
# Using the formula for the sum of a geometric series to calculate the total grains of rice
# Formula: S = a * (r^n - 1) / (r - 1)
# where a is the first term (1 grain), r is the common ratio (2), and n is the number of terms (100).

a = 1  # First term
r = 2  # Common ratio
n = 100  # Number of terms

# Total grains of rice received in 100 days
total_grains = a * (r**n - 1) // (r - 1)

# Output the result
total_grains
```

```python
import matplotlib.pyplot as plt

# Calculate grains of rice per day and cumulative grains of rice
days = range(1, 101)
grains_per_day = [2 ** (day - 1) for day in days]
cumulative_grains = [sum(grains_per_day[:day]) for day in days]

# Create the line graph
plt.figure(figsize=(10, 6))
plt.plot(days, grains_per_day, label="Grains Received on That Day", color="blue", linestyle="-")
plt.plot(days, cumulative_grains, label="Cumulative Grains Received", color="orange", linestyle="--")

# Add labels, title, and legend
plt.xlabel("Days")
plt.ylabel("Number of Grains of Rice")
plt.title("Trend of Grains of Rice Received (Per Day and Cumulative)")
plt.yscale("log")  # Logarithmic scale for better visualization
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Show the graph
plt.tight_layout()
plt.show()
```

```python
# Define a function to calculate the total grains of rice received over a given number of days
def calculate_total_grains(days):
    """
    Calculate the total number of grains of rice received over a specified number of days.
    Each day, the grains double starting with 1 on the first day.

    Args:
    days (int): The number of days.

    Returns:
    int: Total number of grains of rice received.
    """
    return sum(2 ** (day - 1) for day in range(1, days + 1))

# Example usage for 30 and 81 days
grains_for_30_days = calculate_total_grains(30)
grains_for_81_days = calculate_total_grains(81)

grains_for_30_days, grains_for_81_days
```

```python
import matplotlib.pyplot as plt

# Define a function to calculate the total grains of rice for any number of days
def calculate_total_grains(days):
    """
    Calculate the total number of grains of rice received over a specified number of days.
    Each day, the grains double starting with 1 on the first day.

    Args:
    days (int): The number of days.

    Returns:
    int: Total number of grains of rice received.
    """
    return sum(2 ** (day - 1) for day in range(1, days + 1))

# Example usage for 30 and 81 days
days_list = [30, 81]
for days in days_list:
    total_grains = calculate_total_grains(days)
    print(f"Total grains received in {days} days: {total_grains}")

# Generate line graph for a range of days
def plot_grains(days):
    grains_per_day = [2 ** (day - 1) for day in range(1, days + 1)]
    cumulative_grains = [sum(grains_per_day[:day]) for day in range(1, days + 1)]

    # Create the line graph
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, days + 1), grains_per_day, label="Grains Received on That Day", color="blue")
    plt.plot(range(1, days + 1), cumulative_grains, label="Cumulative Grains Received", color="orange", linestyle="--")

    # Add labels, title, and legend
    plt.xlabel("Days")
    plt.ylabel("Number of Grains of Rice")
    plt.title(f"Trend of Grains of Rice Received (1 to {days} Days)")
    plt.yscale("log")  # Logarithmic scale for better visualization
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.tight_layout()
    plt.show()

# Plot grains for 30 days
plot_grains(30)

# Plot grains for 81 days
plot_grains(81)
```

```python
def calculate_days_of_survival(grains_of_rice, num_people):
    """
    Calculate how many days a certain number of people can survive on a given amount of rice.
    
    Args:
    grains_of_rice (int): The total number of grains of rice available.
    num_people (int): The number of people who will be consuming the rice.
    
    Returns:
    int: The number of days the people can survive with the given rice.
    """
    # Based on research, an average person consumes about 5400 grains of rice per day.
    grains_per_person_per_day = 5400
    
    # Total grains of rice needed per day for all people
    total_grains_per_day = grains_per_person_per_day * num_people
    
    # Calculate how many days they can survive
    days_of_survival = grains_of_rice // total_grains_per_day
    
    return days_of_survival

# Example usage: How many days can 10 people survive on 1,000,000 grains of rice?
print(calculate_days_of_survival(1000000, 10))  # Output the result
```
